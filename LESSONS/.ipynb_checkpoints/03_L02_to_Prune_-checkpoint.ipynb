# Our Previous Results

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns


## Reviewing the options used
pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)
pd.set_option('display.float_format', lambda x: f"{x:,.2f}")

## Customization Options
plt.style.use(['fivethirtyeight','seaborn-talk'])
mpl.rcParams['figure.facecolor']='white'

## additional required imports
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import make_column_transformer, make_column_selector, ColumnTransformer
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn import metrics
import joblib
```

## Code/Model From Previous Lesson

```python
## Load in the King's County housing dataset and display the head and info
df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vSEZQEzxja7Hmj5tr5nc52QqBvFQdCAGb52e1FRK1PDT2_TQrS6rY_TR9tjZjKaMbCy1m5217sVmI5q/pub?output=csv")

## Dropping some features for time
df = df.drop(columns=['date'])


## Make the house ids the index
df = df.set_index('id')
\
## drop lat/long
df = df.drop(columns=['lat','long'])
## Treating zipcode as a category
df['zipcode'] = df['zipcode'].astype(str)

df.info()
df.head()
```

```python
# ### Train Test Split
## Make x and y variables
y = df['price'].copy()
X = df.drop(columns=['price']).copy()

## train-test-split with random state for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=321)


# ### Preprocessing + ColumnTransformer

## make categorical & numeric selectors
cat_sel = make_column_selector(dtype_include='object')
num_sel = make_column_selector(dtype_include='number')

## make pipelines for categorical vs numeric data
cat_pipe = make_pipeline(SimpleImputer(strategy='constant',
                                       fill_value='MISSING'),
                         OneHotEncoder(handle_unknown='ignore', sparse=False))

num_pipe = make_pipeline(SimpleImputer(strategy='mean'))

## make the preprocessing column transformer
preprocessor = make_column_transformer((num_pipe, num_sel),
                                       (cat_pipe,cat_sel),
                                      verbose_feature_names_out=False)

## fit column transformer and run get_feature_names_out
preprocessor.fit(X_train)
feature_names = preprocessor.get_feature_names_out()

## Making X_train and X_test into dataframes
X_train_df = pd.DataFrame(preprocessor.transform(X_train), 
                          columns = feature_names, index = X_train.index)
X_train_df.head(3)


X_test_df = pd.DataFrame(preprocessor.transform(X_test), 
                          columns = feature_names, index = X_test.index)
X_test_df.head(3)
```

```python
def evaluate_linreg(model, X_train,y_train, X_test,y_test, get_coeffs=True):
    from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

    results = []
    y_hat_train = model.predict(X_train)
    r2_train = r2_score(y_train,y_hat_train)
    rmse_train = mean_squared_error(y_train,y_hat_train, squared=False)
    results.append({'Data':'Train', 'R^2':r2_train, "RMSE": rmse_train})
    
    y_hat_test = model.predict(X_test)
    r2_test = r2_score(y_test,y_hat_test)
    rmse_test = mean_squared_error(y_test,y_hat_test, squared=False)
    results.append({'Data':'Test', 'R^2':r2_test, "RMSE": rmse_test})
    
    results_df = pd.DataFrame(results).round(3).set_index('Data')
    results_df.index.name=None
    print(results_df)
                    
    if get_coeffs:
        coeffs = pd.Series(model.coef_, index= X_train.columns)
        coeffs.loc['intercept'] = model.intercept_
        return coeffs
```

```python
from sklearn.linear_model import LinearRegression

## fitting a linear regression model
lin_reg = LinearRegression()
lin_reg.fit(X_train_df, y_train)
coeffs = evaluate_linreg(lin_reg, X_train_df, y_train, X_test_df,y_test)
coeffs
```

# Iterating On Our Model

## Removing the Intercept

> First, we can remove the intercept from our model, which will force the LinearRegression to explain all of the price without being free to calculate whatever intercept would help the model.


```python
## fitting a linear regression model
lin_reg = LinearRegression(fit_intercept=False)
lin_reg.fit(X_train_df, y_train)
evaluate_linreg(lin_reg, X_train_df, y_train, X_test_df,y_test)
```

> Removing the intercept did not improve things. Its interesting to note that the coefficients that were very large are STILL very large, but to a larger degree. 

- What is likely hurting our model is all of the zipcode columns. Let's first try removing our zipcodes and see if that improves out model

## Removing Zipcodes

- One approach we could take is to manually remove all zipcode columns from our X variables. For that, we will need to create our list of zip_cols again

```python
## Saving list of zipcode features and other feautures
# Method B) List Comprehension Way
zip_cols = [c for c in X_train_df.columns if c.startswith('zipcode')]
nonzip_cols = [c for c in X_train_df.columns if not c.startswith('zipcode')]


## Preview first 5 zipcols and all nonzip cols
print(zip_cols[:5])        
nonzip_cols
```

```python
X_train_nz = X_train_df.drop(columns=zip_cols)
X_test_nz = X_test_df.drop(columns=zip_cols)
display(X_train_nz.head(3), X_test_nz.head(3))
```

```python
## fitting a linear regression model
lin_reg = LinearRegression(fit_intercept=True)
lin_reg.fit(X_train_nz, y_train)
evaluate_linreg(lin_reg, X_train_nz, y_train, X_test_nz,y_test)
```

> Now those are some better coefficients! The one caveat is that is our model assumed a baseline starting price of $6m. 
- Notice that we had to accept a decrease in performance  (R2 & RMSE) in order to obtain more intuitive coefficients. This may not always be the case, but it will often be the case. 


- Let's remove the intercept and compare our coefficients.

```python
## fitting a linear regression model
lin_reg = LinearRegression(fit_intercept=False)
lin_reg.fit(X_train_nz, y_train)
evaluate_linreg(lin_reg, X_train_nz, y_train, X_test_nz,y_test)
```

## Adding Back to Zipcodes, but in a new form

- While zipcode as a One-Hot-Encoded category was not helpful for our model, there is potentially a way we could use zipcode as an ordinal feature.

- You may be thinking "but the zipcode numbers don't mean anything, how can we replace the categories with integers and still let our model use the information?

- The answer is to rearrange the zipcodes so that they are ranked by average price. This way, the feature has a linear relationship with our target, and our model is still factoring in location in its analysis. 

```python
## calcualte the average price per zipcode and sort values
sorted_zips = df.groupby('zipcode').mean()['price'].sort_values()
sorted_zips.head(3)
```

```python
sorted_zips.plot(kind='bar');
```

- As we can see above, we are able to rearrange the zipcodes be average price, and now we have a clear relationship between the X values and our target.
- We want to replace the zipcode names in the dataframe with integers. You've previously used a similar approach when handling genres in your movie project. 

```python
## Let's make a copy of our original df so we can leave it as-is
df_zips = df.copy()
zip_map = dict(zip(sorted_zips.index, range(len(sorted_zips))))
zip_map.keys()
```

```python
## Replacing zipcodes with integers 
df_zips['zipcode']= df_zips['zipcode'].replace(zip_map)
df_zips
```

```python
## check for linear relationship between new version of zipcpde and price

sns.regplot(data=df_zips, x='zipcode',y='price',
            line_kws={'color':'k'},
           scatter_kws={'ec':'white','lw':0.5})
```

### Modeling with Ordinal Zipcodes

- We will need to remake our X,y, X_train,y_train, etc.
- We will also want to make a copy of our preprocessor so we can leave the original one intact with the values it learned from the earlier X/y data.

```python
## clone a sklearn object with the clone function
from sklearn.base import clone
preprocessor_zips = clone(preprocessor)
preprocessor_zips
```

```python
#
## fit column transformer and run get_feature_names_out
## Make x and y variables
y = df_zips['price'].copy()
X = df_zips.drop(columns=['price']).copy()

## train-test-split with random state for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=321)



preprocessor_zips.fit(X_train)
feature_names_zips = preprocessor_zips.get_feature_names_out()


## Making X_train and X_test into dataframes
X_train_zips = pd.DataFrame(preprocessor_zips.transform(X_train), 
                          columns = feature_names_zips, index = X_train.index)


X_test_zips = pd.DataFrame(preprocessor_zips.transform(X_test), 
                          columns = feature_names, index = X_test.index)
display(X_train_df.head(3),X_test_df.head(3))
```

```python
## fitting a linear regression model
lin_reg = LinearRegression(fit_intercept=True)
lin_reg.fit(X_train_zips, y_train)
coeffs_zips_int = evaluate_linreg(lin_reg, X_train_zips, y_train, X_test_zips,y_test)
coeffs_zips_int
```

> Comparing to a model without an intercept

```python
## fitting a linear regression model
lin_reg = LinearRegression(fit_intercept=False)
lin_reg.fit(X_train_zips, y_train)
coeffs_zips_noint = evaluate_linreg(lin_reg, X_train_zips, y_train, X_test_zips,y_test)
coeffs_zips_noint
```

### To Intercept or Not To Intercept?

- At this point, there is a valid argument for using either model as the basis for our stakeholder recommendations. 
- As long as you are comfortable explaining the intercept as the baseline house price (when all Xs are 0), then it is not difficult to express the findings to a stakeholder.
- Let's see if either version looks better when visualzied.


```python
fig, axes = plt.subplots(ncols=2,figsize=(12,8))

coeffs_zips_int.plot(kind='barh',color='blue',ax=axes[0],title='Coefficients + Intercept')
coeffs_zips_noint.plot(kind='barh',color='green',ax=axes[1],title='Coefficients - No intercept')
fig.tight_layout()
```

> Visually, it is difficult to compare coefficients when they are based on the raw features. The effect of increase 1 sqft of living space is going to MUCH smaller than the effect of having waterfront property. 

- At this point, we may want to consider scaling our features. It will complicate the non-technical interpretation a little bit, but it may give us more helpful/impactful visualizations.

## Scaling Our Features

<font color=red>Question: wanted originally to show making multiple preprocessing pipelines by just changing the numeric pipeline. Switched to adding a scaler AFTER preprocessing. My rationale was that since all of the features are numeric at this pont, there is really no point in recreating the whole preprocessor. </font>

- Since we have entirely numeric features, we can simply scale our already-processed X_train/X_test variables by creating a new scaler. 
    - Note: for more complicated datasets, we would want to create a new precprocessor where we add the scaler to the numeric pipeline.

```python
scaler = StandardScaler()
X_train_zips_scaled = pd.DataFrame(scaler.fit_transform(X_train_zips),
                                   columns=X_train_zips.columns, 
                                   index= X_train_zips.index)

X_test_zips_scaled = pd.DataFrame(scaler.transform(X_test_zips),         
                                      columns=X_test_zips.columns, 
                                      index= X_test_zips.index)

X_train_zips_scaled.head()
```

```python
## fitting a linear regression model
lin_reg = LinearRegression(fit_intercept=False)
lin_reg.fit(X_train_zips_scaled, y_train)
coeffs_zips_scaled = evaluate_linreg(lin_reg, X_train_zips_scaled, y_train, X_test_zips_scaled,y_test)
coeffs_zips_scaled
```
